```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Specify the path to your downloaded point shapefile
point_shapefile_path = r'C:\Users\Subhadeep Maishal\Music\JUP\cyclone_data\cyclone_data\IBTrACS.ALL.list.v04r00.points.shp'

# Read the point shapefile into a GeoDataFrame
gdf_points = gpd.read_file(point_shapefile_path)

# Specify the path to the coastline shapefile
coastline_shapefile_path = r'D:\ne_10m_coastline\ne_10m_coastline.shp'

# Read the coastline shapefile into a GeoDataFrame
gdf_coastline = gpd.read_file(coastline_shapefile_path)

# Define the latitude and longitude box (example: for the region of interest)
min_lat, max_lat = 0, 20  # Define your latitude range
min_lon, max_lon = 60, 80  # Define your longitude range

# Subset the GeoDataFrames based on the latitude and longitude box
gdf_points_subset = gdf_points[(gdf_points['LATITUDE'] >= min_lat) & (gdf_points['LATITUDE'] <= max_lat)
                               & (gdf_points['LONGITUDE'] >= min_lon) & (gdf_points['LONGITUDE'] <= max_lon)]

gdf_coastline_subset = gdf_coastline.cx[min_lon:max_lon, min_lat:max_lat]

# Plot the coastline
ax = gdf_coastline_subset.plot(figsize=(12, 8), color='lightgray')

# Plot the cyclone points on top of the coastline
gdf_points_subset.plot(ax=ax, color='red', markersize=10)

plt.title('Cyclone Points with Coastline (Subset)')
plt.show()
```

```python
from mpi4py import MPI
import geopandas as gpd
import matplotlib.pyplot as plt

def plot_subset_and_coastline(comm, rank, size, min_lat, max_lat, min_lon, max_lon):
    # Read the point shapefile into a GeoDataFrame
    gdf_points = gpd.read_file(r'C:\Users\Subhadeep Maishal\Music\JUP\cyclone_data\cyclone_data\IBTrACS.ALL.list.v04r00.points.shp')

    # Read the coastline shapefile into a GeoDataFrame
    gdf_coastline = gpd.read_file(r'D:\ne_10m_coastline\ne_10m_coastline.shp')

    # Calculate the number of points for each process
    points_per_process = len(gdf_points) // size
    remainder = len(gdf_points) % size

    # Distribute the workload evenly among processes
    start_idx = rank * points_per_process
    end_idx = (rank + 1) * points_per_process if rank < size - 1 else (rank + 1) * points_per_process + remainder

    # Subset the GeoDataFrame based on the assigned workload
    gdf_points_subset = gdf_points.iloc[start_idx:end_idx]

    # Perform the subset operation collectively across all processes
    gdf_points_subset = comm.gather(gdf_points_subset, root=0)

    if rank == 0:
        # Concatenate the results from all processes
        gdf_points_subset = gpd.GeoDataFrame(pd.concat(gdf_points_subset, ignore_index=True), crs=gdf_points.crs)

        # Subset the coastline GeoDataFrame
        gdf_coastline_subset = gdf_coastline.cx[min_lon:max_lon, min_lat:max_lat]

        # Plot the coastline
        ax = gdf_coastline_subset.plot(figsize=(12, 8), color='lightgray')

        # Plot the cyclone points on top of the coastline
        gdf_points_subset.plot(ax=ax, color='red', markersize=10)

        plt.title('Cyclone Points with Coastline (Parallel)')
        plt.show()

if __name__ == "__main__":
    comm = MPI.COMM_WORLD
    rank = comm.Get_rank()
    size = comm.Get_size()

    # Define the latitude and longitude box
    min_lat, max_lat = 0, 20  # Define your latitude range
    min_lon, max_lon = 60, 80  # Define your longitude range

    # Perform parallel processing
    plot_subset_and_coastline(comm, rank, size, min_lat, max_lat, min_lon, max_lon)
```
